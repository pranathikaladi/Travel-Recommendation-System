```python
import pandas as pd
```

```python
#remove spaces from the category column of dataset
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
metadata = pd.read_csv('1.csv')
metadata.head()
```

```python
metadata['types_of_travel'] = metadata['types_of_travel'].apply(clean_data)
print("Select your preferred category:\n1.solo \n2.family \n3.friends")
text = input("Enter User Interests: ")   #user preference
text='travel_type_'+text
pref1=metadata[text]==1
pref2=metadata['travel_type_any']==1
filter1=metadata[pref1].append(metadata[pref2])
filter1=filter1.sort_values(by='location_rating',ascending=False)
filter1[['location','place','state','location_rating']].head()
```

```python
metadata['season'] = metadata['season'].apply(clean_data)
print("Select your preferred category:\n1.autumn \n2.rainy \n3.spring \n4.summer \n5.winter")
inp = input("Enter Season:")
inp1='season_'+inp
input1=metadata[inp1]==1
input2=metadata['season_any']==1
filter2=metadata[input1].append(metadata[input2])
filter2=filter2.sort_values(by='location_rating',ascending=False)
filter2[['location','place','state','location_rating']].head()
```

```python
user_age = int(input("Enter Age:"))
any_age_people=metadata.age == 0
if(user_age<15):
    k = metadata.age <= 15
    filter3=metadata[k]
elif(user_age>50):
    l = metadata.age == 50
    filter3=metadata[l].append(metadata[any_age_people])
else:
    a = (user_age-10) < metadata.age
    b = metadata.age <= (user_age+10)
    f1=metadata[a]
    f2=metadata[b]
    f3=pd.merge(f1, f2, how ='inner')
    filter3=f3.append(metadata[any_age_people])
filter3=filter3.sort_values(by='location_rating',ascending=False)
filter3[['location','place','state','location_rating']].head()
```

```python
metadata['state'] = metadata['state'].apply(clean_data)
inp_state = input("Enter the state:")#user preference
st = metadata.state == inp_state
filter4=metadata[st]
filter4=filter4.sort_values(by='location_rating',ascending=False)
filter4[['location','place','state','location_rating']].head()
```

```python
metadata['religion'] = metadata['religion'].apply(clean_data)
print("Select your preferred category:\n1.hindus \n2.muslim \n3.jain \n4.buddist \n5.sikhs \n6.christian")
rel = input("Enter Religion:")
rel1='religion_'+rel
input1=metadata[rel1]==1
filter5=metadata[input1]
filter5=filter5.sort_values(by='location_rating',ascending=False)
filter5[['location','place','state','location_rating']].head()
```

```python
metadata['Category'] = metadata['Category'].apply(clean_data)
print("Select your preferred category:\n1.pilgrimage \n2.wildlife \n3.waterfalls \n4.park \n5.water \n6.historic \n7.museum \n8.desert \n9.valley \n10.hill station \n11.industrial")
cat = input("Enter category:")
cat1=metadata.Category == cat
filter6=metadata[cat1]
filter6=filter6.sort_values(by='location_rating',ascending=False)
filter6[['location','place','state','location_rating']].head()
```
